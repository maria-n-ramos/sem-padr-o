**Calculando o Primeiro Dígito Verificador**

O primeiro dígito verificador do CPF é calculado utilizando-se o seguinte algoritmo:

1) Multiplique os valores de cada coluna e some:


*x* 10	9	8	7	6	5	4	3	2


```python
#separar os digitos
lista_digito = [int(a) for a in str(cpf)]
#só valem os 9 primeiros
#vamos multiplicar por 10, 9, 8, 7, 6, 5, 4, 3, 2 da esquerda para a direita
contador = 0
soma = 0
vezes = 10
while contador <= 8:
  multiplica =  lista_digito[contador] * (vezes)
  soma = soma + multiplica
  contador = contador + 1
  vezes = vezes - 1
```

4) O resultado obtido será divido por 11. Considere como quociente apenas o valor inteiro, o resto da divisão será responsável pelo cálculo do primeiro dígito verificador.

```python
primeiro_digito = 11 - (soma % 11)
print(primeiro_digito)
```

**Calculando o Segundo Dígito Verificador**

1) Para o cálculo do segundo dígito será usado o primeiro dígito verificador já calculado. Dessa vez multiplicaremos os valores por 11,10,9,8,7,6,5,4,3,2 já que estamos incorporando mais um algarismo para esse cálculo e efetuaremos o somatório dos resultados obtidos.

```python
lista_digito.append(1)

contador = 0
soma = 0
j = 11
while contador <= 9:
  multiplica =  lista_digito[contador] * (j)
  soma = soma + multiplica
  contador = contador + 1
  j = j - 1
  print(multiplica, soma)
```

Realizamos novamente o cálculo do módulo 11. Dividimos o total do somatório por 11 e consideramos o resto da divisão.
Caso o valor do resto da divisão seja menor que 2, esse valor passa automaticamente a ser zero, caso contrário (como no nosso caso) é necessário subtrair o valor obtido de 11 para se obter o dígito verificador.

```python
segundo_digito = 11 - (soma % 11)
print(segundo_digito)
```


# **VERIFICANDO OS DOIS ÚLTIMOS DÍGITOS**
