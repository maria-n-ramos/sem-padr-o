```python
def digitos_verificadores(lista_digito):
    #só valem os 9 primeiros
    #vamos multiplicar por 10, 9, 8, 7, 6, 5, 4, 3, 2 da esquerda para a direita
    contador = 0
    soma = 0
    vezes = 10
    while contador <= 8:
      multiplica =  lista_digito[contador] * (vezes)
      #pint("dig",lista_digito[contador],"vzs",vezes) verifica se a conta está de acordo
      soma = soma + multiplica
      contador = contador + 1
      vezes -=1
    subtraendo = soma % 11
    primeiro_digito = 11 - (soma % 11)
    if (subtraendo) < 1:
        primeiro_digito = 0

#cálculo do segundo dígito verificador e dessa vez multiplicaremos os valores por
#11,10,9,8,7,6,5,4,3,2 já que estamos incorporando mais um algarismo para esse cálculo 
#e efetuaremos o somatório dos resultados obtidos.  
    lista_digito.append(1)

    contador = 0
    soma = 0
    j = 11
    while contador <= 9:
      multiplica =  lista_digito[contador] * (j)
      soma = soma + multiplica
      contador = contador + 1
      j = j - 1
    subtraendo = soma % 11
    segundo_digito = 11 - (subtraendo)
    if (subtraendo) < 1:
        segundo_digito = 0
        
    lista_digito.pop() #retirando o extra dígito 1
    return primeiro_digito,segundo_digito
```
